# Importing Credit Card details 

```python
import pandas as pd
from sqlalchemy import create_engine
import os
from urllib.parse import quote_plus


# File path and database credentials
csv_file_path = "C:/Users/aksha/Desktop/Credit Card Power BI project/credit_card.csv"
username = "root"
password = "Aksh@y_mysql"
host = "localhost"
database = "cc_db"

# Load CSV into Pandas DataFrame
data = pd.read_csv(csv_file_path)
print("CSV Data Preview:")
print(data.head())

# MySQL credentials
username = "root"  # MySQL username
password = "Aksh@y_mysql"  # MySQL password
host = "localhost"  # MySQL server host
database = "cc_db"  # MySQL database name

# URL-encode the password
encoded_password = quote_plus(password)

# Properly formatted connection string
connection_string = f"mysql+pymysql://{username}:{encoded_password}@{host}/{database}"

# Create the SQLAlchemy engine
engine = create_engine(connection_string)

# Load the CSV file
csv_file_path = "C:/Users/aksha/Desktop/Credit Card Power BI project/credit_card.csv" # Replace with your actual file path
data = pd.read_csv(csv_file_path)

# Import DataFrame into MySQL
try:
    data.to_sql(name='cc_detail', con=engine, if_exists='append', index=False)
    print("Data imported successfully!")
except Exception as e:
    print(f"Error during import: {e}")
```
